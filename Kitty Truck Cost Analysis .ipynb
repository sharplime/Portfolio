```python
# Import necessary libraries. 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Suppress warnings. 
import warnings  
warnings.filterwarnings('ignore')
```

```python
# Import the CSV file.
kite = pd.read_excel('miles50_coventry.xlsx')
```

```python
# View the DataFrame.
print(kite.shape)
kite.head()
```

```python
# Drop first column. 
kite = kite.drop('Unnamed: 0', axis=1)
```

```python
# View the DataFrame.
print(kite.shape)
kite.head()
```

```python
# Explore the metadata.
kite.info()
```

```python
# Convert the 'OrderDate' column to datetime type
kite['OrderDate'] = pd.to_datetime(kite['OrderDate'])

# Remove the time component and keep only the date
kite['OrderDate'] = kite['OrderDate'].dt.date

# Display the updated DataFrame
kite.head()
```

### View the number of pallets by each courier. 

```python
# Pivot the data to see orders by day by courier. 
df_pallets_by_day = kite.pivot_table(index='Courier', columns='OrderDate', values='Pallet_per_line', aggfunc='sum')

# Display the DataFrame
df_pallets_by_day
```

```python
# Remove DX Express as will only look to replace DX Freight, MFS, and Spit orders. 
courier_to_remove = 'DX Express'

# Drop the row corresponding to the specified courier
df_pallets_by_day = df_pallets_by_day.drop(courier_to_remove)

# Display the updated DataFrame
print(df_pallets_by_day)
```

```python
# Calculate the average pallet per day for each courier
average_pallet_per_day = df_pallets_by_day.mean(axis=1)

# Display the average pallet per day for each courier
average_pallet_per_day
```

```python
# Define the pallet capacities of the trucks
pallet_capacity_3_5_tonne_van = 2
pallet_capacity_7_5_tonne_LGV = 14
pallet_capacity_18_tonne_LGV = 24

# Create a new DataFrame to store the truck type and number of trucks needed for each courier on average.
df_trucks_needed_avg = pd.DataFrame(index=average_pallet_per_day.index)

# Add columns for average pallet count and each truck size
df_trucks_needed_avg['Average_Pallets_Day'] = average_pallet_per_day
df_trucks_needed_avg['3.5-tonne Van (2)'] = 0
df_trucks_needed_avg['7.5-tonne LGV (14)'] = 0
df_trucks_needed_avg['18-tonne LGV (24)'] = 0

df_trucks_needed_avg
```

```python
# Use For loop to determine the truck type and number of trucks needed for each courier.
# Based on the average number of pallets needed per day.

for courier in df_trucks_needed_avg.index:
    pallets_needed = df_trucks_needed_avg.loc[courier, 'Average_Pallets_Day']

    # Check if the number of pallets needed is >= to the capacity of the largest truck size (18-tonne LGV). 
    if pallets_needed >= pallet_capacity_18_tonne_LGV:
        trucks_needed_18_tonne = pallets_needed // pallet_capacity_18_tonne_LGV
        pallets_needed %= pallet_capacity_18_tonne_LGV
        df_trucks_needed_avg.loc[courier, '18-tonne LGV (24)'] = trucks_needed_18_tonne

    # Check if the number of pallets needed is >= to the capacity of the 7.5-tonne LGV. 
    if pallets_needed >= pallet_capacity_7_5_tonne_LGV:
        trucks_needed_7_5_tonne = pallets_needed // pallet_capacity_7_5_tonne_LGV
        pallets_needed %= pallet_capacity_7_5_tonne_LGV
        df_trucks_needed_avg.loc[courier, '7.5-tonne LGV (14)'] = trucks_needed_7_5_tonne

    # Check if the number of pallets needed is >= to the capacity of the 3.5-tonne van. 
    if pallets_needed >= pallet_capacity_3_5_tonne_van:
        trucks_needed_3_5_tonne = pallets_needed // pallet_capacity_3_5_tonne_van
        pallets_needed %= pallet_capacity_3_5_tonne_van
        df_trucks_needed_avg.loc[courier, '3.5-tonne Van (2)'] = trucks_needed_3_5_tonne
        
# Add a total row that sums the number of trucks needed for each truck type
df_trucks_needed_avg.loc['Total'] = df_trucks_needed_avg.sum()

# Display the DataFrame with the truck type and number of trucks needed for each courier on average
df_trucks_needed_avg
```

```python
# Define the monthly cost for each truck type
monthly_cost_3_5_tonne_van = 3400
monthly_cost_7_5_tonne_LGV = 4550
monthly_cost_18_tonne_LGV = 5400

# Create a new row with the monthly cost multiplied by the number of trucks needed for each truck type
df_monthly_cost = pd.DataFrame({
    'Average_Pallets_Day': np.nan,
    '3.5-tonne Van (2)': df_trucks_needed_avg.loc['Total', '3.5-tonne Van (2)'] * monthly_cost_3_5_tonne_van,
    '7.5-tonne LGV (14)': df_trucks_needed_avg.loc['Total', '7.5-tonne LGV (14)'] * monthly_cost_7_5_tonne_LGV,
    '18-tonne LGV (24)': df_trucks_needed_avg.loc['Total', '18-tonne LGV (24)'] * monthly_cost_18_tonne_LGV
}, index=['Monthly_Cost'])

# Concatenate the new row with the original DataFrame
df_trucks_needed_avg_with_cost = pd.concat([df_trucks_needed_avg, df_monthly_cost])

# Display the updated DataFrame with the truck type, number of trucks needed, and monthly cost for each courier on average
df_trucks_needed_avg_with_cost
```

```python
# Sum the total monthly truck costs. 
monthly_truckcost = df_trucks_needed_avg_with_cost.loc['Monthly_Cost'].sum()
monthly_truckcost
```

```python
# See the total delivery cost for Kite. 
total_delivery_cost = kite['DeliveryCost'].sum()
total_delivery_cost
```

```python
# Work out estimated delivery costs per month. 
# Get the number of unique days.
num_days = len(kite['OrderDate'].unique()) 
num_days
```

```python
# Get the sum of the 'DeliveryCost' column.
total_delivery_cost = kite['DeliveryCost'].sum() 

# Calculate the average cost per day.
average_cost_per_day = total_delivery_cost / num_days

# Calculate the average cost per month.
# Numbers of days in a month without weekends or bank holidays = 22. 
average_delivery_per_month = average_cost_per_day *22
average_delivery_per_month
```

```python
# Create a new DataFrame with the total monthly cost and total delivery cost.
df_costs_summary = pd.DataFrame({
    'Truck_cost_month': monthly_truckcost,
    'Delivery_Cost_month': average_delivery_per_month
}, index=['Summary'])

# Round the values in the DataFrame to two decimal places.
df_costs_summary = df_costs_summary.round(2)

# Display the new DataFrame.
df_costs_summary
```

```python
# Add a new column showing the percentage savings. 
df_costs_summary['Savings%'] = (df_costs_summary['Truck_cost_month'] / df_costs_summary['Delivery_Cost_month']) * 100

# Round the values in the "Savings%" column to 2 decimal places.
df_costs_summary['Savings%'] = df_costs_summary['Savings%'].round(2)

# Display the new DataFrame.
df_costs_summary
```
