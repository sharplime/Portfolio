```python
# Import necessary libraries. 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Suppressing warnings
import warnings  
warnings.filterwarnings('ignore')
```

```python
# Import the CSV file.
kite = pd.read_csv('clustered_data.csv')
```

```python
# View the DataFrame.
print(kite.shape)
kite.head()
```

```python
# Drop first column. 
kite = kite.drop('Unnamed: 0', axis=1)
```

```python
# View the DataFrame.
print(kite.shape)
kite.head()
```

```python
# Explore the metadata.
kite.info()
```

```python
# Convert the 'OrderDate' column to datetime type
kite['OrderDate'] = pd.to_datetime(kite['OrderDate'])

# Remove the time component and keep only the date
kite['OrderDate'] = kite['OrderDate'].dt.date

# Convert 'DateDespatched' column to datetime format
kite['DateDespatched'] = pd.to_datetime(kite['DateDespatched'])

# Display the updated DataFrame
kite.head()
```

### Only Looking at Working Days

```python
# Creating a column to identify the day of the week that the order was dispatched
kite['weekday'] = kite['DateDespatched'].apply(lambda x: x.weekday())

# Removing orders from the DataFrame that were dispatched on a Saturday (5) or Sunday (6)    
kite_wd = kite.loc[~((kite['weekday'] == 5) | (kite['weekday'] == 6))]

# Checking this has worked
kite_wd['weekday'].unique()
```

```python
# Removing orders that were dispatched on the Easter bank holidays April 7th & 10th
kite_wd = kite_wd[~kite_wd['DateDespatched'].isin([pd.Timestamp('20230407'), pd.Timestamp('20230410')])]

# Checking this has worked
kite_wd['DateDespatched'].unique()
```

```python
# Viewing the DataFrame
kite_wd.head()
```

```python
# View the DataFrame.
print(kite_wd.shape)
kite_wd.head()
```

### View the number of pallets by each courier. 

```python
# Pivot the data to see orders by day by courier. 
df_pallets_by_day = kite_wd.pivot_table(index='Courier', columns='DateDespatched', values='Pallets', aggfunc='sum')

# Display the DataFrame
df_pallets_by_day
```

```python
# Remove DX Express as will only look to replace DX Freight, MFS, and Spit orders. 
courier_to_remove = 'DX Express'

# Drop the row corresponding to the specified courier
df_pallets_by_day = df_pallets_by_day.drop(courier_to_remove)

# Display the updated DataFrame
print(df_pallets_by_day)
```

```python
# Calculate the average pallet per day for each courier
average_pallet_per_day = df_pallets_by_day.median(axis=1)  

average_pallet_per_day 
```

### View the number of orders by each courier. 

```python
# Pivot the data to see number of orders by day by courier. 
df_orders_by_day = kite_wd.pivot_table(index='Courier', columns='DateDespatched', values='OrderID', aggfunc='count')

# Display the DataFrame
df_orders_by_day
```

```python
# Calculate the average orders per day for each courier
average_order_per_day = df_orders_by_day.median(axis=1)  

average_order_per_day
```

### Work out optimal truck capacity for daily pallet average 

#### Assumption that we would prefer to fill a larger vehicle to 70% rather than have multiple smaller trucks. 

```python
import math

pallet_capacity_3_5_tonne_van = 2
pallet_capacity_7_5_tonne_LGV = 14
pallet_capacity_18_tonne_LGV = 24
fill_capacity_percent = 0.7

# Create a new DataFrame to store the truck type and number of trucks needed for each courier on average.
df_trucks_needed_avg = pd.DataFrame(index=average_pallet_per_day.index)

# Add columns for average pallet count and each truck size
df_trucks_needed_avg['Average_Orders_Day'] = average_order_per_day
df_trucks_needed_avg['Average_Pallets_Day'] = average_pallet_per_day
df_trucks_needed_avg['3.5-tonne Van (2)'] = 0
df_trucks_needed_avg['7.5-tonne LGV (14)'] = 0
df_trucks_needed_avg['18-tonne LGV (24)'] = 0
```

```python
# Use a for loop to determine the truck type and number of trucks needed for each courier
for courier in df_trucks_needed_avg.index:
    pallets_needed = df_trucks_needed_avg.loc[courier, 'Average_Pallets_Day']

    # Check if the number of pallets needed is greater than or equal to the capacity of the 18-tonne LGV
    if pallets_needed >= math.ceil(pallet_capacity_18_tonne_LGV * fill_capacity_percent):
        trucks_needed_18_tonne = math.ceil(pallets_needed / pallet_capacity_18_tonne_LGV)
        df_trucks_needed_avg.loc[courier, '18-tonne LGV (24)'] = trucks_needed_18_tonne

    # Check if the number of pallets needed is greater than or equal to the capacity of the 7.5-tonne LGV
    elif pallets_needed >= math.ceil(pallet_capacity_7_5_tonne_LGV * fill_capacity_percent):
        trucks_needed_7_5_tonne = math.ceil(pallets_needed / pallet_capacity_7_5_tonne_LGV)
        df_trucks_needed_avg.loc[courier, '7.5-tonne LGV (14)'] = trucks_needed_7_5_tonne

    # If the number of pallets needed is less than the fill capacity of the 7.5-tonne LGV, use smaller trucks
    else:
        trucks_needed_3_5_tonne = math.ceil(pallets_needed / pallet_capacity_3_5_tonne_van)
        df_trucks_needed_avg.loc[courier, '3.5-tonne Van (2)'] = trucks_needed_3_5_tonne

# Add a total row that sums the number of trucks needed for each truck type
df_trucks_needed_avg.loc['Total'] = df_trucks_needed_avg.sum()
        
# Display the DataFrame with the truck type and number of trucks needed for each courier on average
df_trucks_needed_avg
```

### Work out optimal truck capacity for maximum of 25 orders a day

```python
# Copy the Dataframe. 
df_trucks_25_max = df_trucks_needed_avg.copy()

# Drop the Total row. 
df_trucks_25_max = df_trucks_25_max.drop('Total')
```

```python
# Calculated a % of pallets based on the % of 25 max orders / Average_Orders_Day.
# Ignored if Average_Orders_Day was less than 25. 

# Calculate the order percentage by dividing 25 by the average orders per day
df_trucks_25_max['Percentage_25max_orders'] = 25 / df_trucks_25_max['Average_Orders_Day']

# Apply the order percentage to the average pallets per day, ignoring values where Average_Orders_Day < 25
df_trucks_25_max['Pallets_25max_orders'] = df_trucks_25_max.apply(lambda row: row['Average_Pallets_Day'] 
                                                                      * row['Percentage_25max_orders'] 
                                                                      if row['Average_Orders_Day'] >= 25 
                                                                      else row['Average_Pallets_Day'], axis=1)

df_trucks_25_max
```

```python
# Specify the columns you want to include in the new DataFrame
columns_to_include = ['Pallets_25max_orders']

# Create the new DataFrame with the selected columns
df_trucks_25_max = df_trucks_25_max[columns_to_include]

df_trucks_25_max
```

```python
# Work out new trucks needed based on this daily pallet number for 25 max orders. 

import math

pallet_capacity_3_5_tonne_van = 2
pallet_capacity_7_5_tonne_LGV = 14
pallet_capacity_18_tonne_LGV = 24
fill_capacity_percent = 0.7

# Add columns for average pallet count and each truck size
df_trucks_25_max['3.5-tonne Van (2)'] = 0
df_trucks_25_max['7.5-tonne LGV (14)'] = 0
df_trucks_25_max['18-tonne LGV (24)'] = 0

df_trucks_25_max
```

```python
# Use a for loop to determine the truck type and number of trucks needed for each courier
for courier in df_trucks_25_max.index:
    pallets_needed = df_trucks_25_max.loc[courier, 'Pallets_25max_orders']

    # Check if the number of pallets needed is greater than or equal to the capacity of the 18-tonne LGV
    if pallets_needed >= math.ceil(pallet_capacity_18_tonne_LGV * fill_capacity_percent):
        trucks_needed_18_tonne = 1
        trucks_needed_7_5_tonne = 1
        df_trucks_25_max.loc[courier, '18-tonne LGV (24)'] = trucks_needed_18_tonne
        df_trucks_25_max.loc[courier, '7.5-tonne LGV (14)'] = trucks_needed_7_5_tonne

    # Check if the number of pallets needed is greater than or equal to the capacity of the 7.5-tonne LGV
    elif pallets_needed >= math.ceil(pallet_capacity_7_5_tonne_LGV * fill_capacity_percent):
        trucks_needed_7_5_tonne = 1
        df_trucks_25_max.loc[courier, '7.5-tonne LGV (14)'] = trucks_needed_7_5_tonne

    # If the number of pallets needed is less than the capacity of the 7.5-tonne LGV, use a smaller truck
    else:
        trucks_needed_3_5_tonne = 1
        df_trucks_25_max.loc[courier, '3.5-tonne Van (2)'] = trucks_needed_3_5_tonne

# Add a total row that sums the number of trucks needed for each truck type
df_trucks_25_max.loc['Total'] = df_trucks_25_max.sum()
        
# Display the updated DataFrame
df_trucks_25_max
```
